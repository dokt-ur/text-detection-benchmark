```python
import os
from glob import glob 
from uuid import uuid4

import cv2
from matplotlib import pyplot as plt

from paddle_onnx import PaddleOnnx
```

```python
paddle_onnx = PaddleOnnx()
```

```python
img = cv2.imread(paddle_onnx.test_image_path)
```

```python
output = paddle_onnx.process_image(img)
```

```python
plt.imshow(output)
```

```python
OUTPUT_DIR = "output"
transaction_id = str(uuid4())
target_dir = f"{OUTPUT_DIR}/{transaction_id}"

if not os.path.exists(target_dir):
    os.makedirs(target_dir)

text_dir = "../benchmark/imgs/test-set/text/resized/0.5"

text_file_list = glob(f"{text_dir}/*.jpg")
paddle_onnx.run_inference(text_file_list, target_dir)

print(f"Output images can be found at {target_dir}")
```
